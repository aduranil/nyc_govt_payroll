```python
import pandas as pd
```

```python
from sodapy import Socrata
```

```python
client = Socrata("data.cityofnewyork.us", "1vscl5xBtxiCI6ReKRmVwa0zt")
```

```python
final_results = []
loops = 67 
records = 50000
identified = "k397-673e"
for i in range(loops):
    results = client.get(identified, limit=records, offset=records*i)
    final_results.append(pd.DataFrame(results))
```

```python
updated_result = pd.concat(final_results, sort=True)
```

```python
def filter_result_by_year(result, year):
    return result[result['fiscal_year'] == year]
```

```python
filter_result_by_year(updated_result, "2019")
```

```python
def filter_column_by_value(result, column, value):
    return result[result[column].str.contains(value)]
```

```python
yearly_result = filter_result_by_year(updated_result, "2019")
```

```python
annual_result = filter_column_by_value(yearly_result, 'pay_basis', 'Annum')
```

```python
annual_result
```

```python
def add_count_of_value(result, new_column_name, value_count_column):
    result[new_column_name] = result[value_count_column].map(result[value_count_column].value_counts())
    return result
```

```python
active_result = filter_column_by_value(annual_result, 'leave_status_as_of_july_31', 'ACTIVE')
```

```python
active_result
```

```python
add_count_of_value(active_result, 'name_count', 'first_name')
```

```python
import numpy as np
```

```python
def get_data_set(result, column, threshold, groupby_col, agg_call):
    return result[result[column] > threshold].groupby(groupby_col).agg({agg_call: np.mean}).sort_values(agg_call, ascending=False)
```

```python
get_data_set(active_result, 'name_count', 1000, 'first_name', 'base_salary')
```

```python
updated_active_result = add_count_of_value(annual_result, 'name_count', 'first_name')
```

```python
get_data_set(updated_active_result, 'name_count', 1000, 'first_name', 'base_salary')
```
