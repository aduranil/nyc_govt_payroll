```python
from sodapy import Socrata
```

```python
client = Socrata("data.cityofnewyork.us", "1vscl5xBtxiCI6ReKRmVwa0zt")
```

```python
final_results = []
loops = 67 
records = 50000
identified = "k397-673e"
for i in range(loops):
    results = client.get(identified, limit=records, offset=records*i)
    final_results.append(pd.DataFrame(results))
```

```python
def filter_result_by_year(result, year):
    return result[result['fiscal_year'] == year]
```

```python
filter_result_by_year(updated_result, "2019")
```

```python
def filter_column_by_value(result, column, value):
    return result[result[column].str.contains(value)]
```

```python
yearly_result = filter_result_by_year(updated_result, "2019")
```

```python
annual_result = filter_column_by_value(yearly_result, 'pay_basis', 'Annum')
```

```python
def add_count_of_value(result, new_column_name, value_count_column):
    result[new_column_name] = result[value_count_column].map(result[value_count_column].value_counts())
    return result
```

```python
active_result = filter_column_by_value(annual_result, 'leave_status_as_of_july_31', 'ACTIVE')
```

```python
add_count_of_value(active_result, 'name_count', 'first_name')
```

```python
import numpy as np
```

```python
def get_data_set(result, column, threshold, groupby_col, agg_call):
    return result[result[column] > threshold].groupby([groupby_col, 'name_count']).agg({agg_call: np.mean}).sort_values(agg_call, ascending=False).reset_index()
```

```python
get_data_set(active_result, 'name_count', 1000, 'first_name', 'base_salary')
```

```python
def filter_greater_than(result, column, threshold):
    return result[result[column] > threshold]
```

```python
def filter_all(result, year, threshold):
    updated_result = filter_result_by_year(result, year)
    pay_basis_result = filter_column_by_value(updated_result, 'pay_basis', 'Hour')
    lesser_result = filter_less_than(pay_basis_result, 'base_salary', 20000)
    add_count_of_value(lesser_result, 'name_count', 'first_name')
    return get_data_set(lesser_result, 'name_count', threshold, 'first_name', 'base_salary')
    # get_data_set(lesser_result, 'first_name', threshold, 'first_name', 'base_salary')
```

```python
filter_all(updated_result, '2019', 150)
```

```python
def filter_all_salary(result, threshold):
    updated_result = filter_result_by_year(result, '2019')
    greater_result = filter_column_by_value(updated_result, 'pay_basis', 'Annum')
#     greater_result = filter_greater_than(pay_basis_result, 'base_salary', 1)
#     lesser_result = filter_less_than(greater_result, 'base_salary', 20000)
    add_count_of_value(greater_result, 'name_count', 'first_name')
    return get_data_set(greater_result, 'name_count', threshold, 'first_name', 'base_salary')
    # get_data_set(lesser_result, 'first_name', threshold, 'first_name', 'base_salary')
```
